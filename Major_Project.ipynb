**IMPORT LIBRARIES AND DATASETS**

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/suv_data.csv')
df
```

**PERFORMING EDA**

```python
df.info()
```

```python
x = df.iloc[:,[2,3]].values
x
```

```python
y = df.iloc[:,4].values
y
```

```python
df.head()
```

```python
gender = pd.get_dummies(df['Gender'], drop_first = True)
gender.head()
```

```python
Data = pd.concat([df , gender ],axis = 1)
Data.head()                 
```

**DIVIDE DATA INTO INPUT AND OUTPT**

```python
x = Data.iloc[:,[2,3,5]].values
y = Data.iloc[:,4].values

x[:10],y[:10]
```

**SPLITING DATA INTO TRAINING ANS TESTING SET**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

**NORMALISING/SCALING THE INPUT VALUES**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

**APPLYING THE LOGISTIC REGRESSOR**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

**FITTING THE MODEL**

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

**MODEL EVALUATION**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

**SO THE ACCURACY SCORE IS 91%**
